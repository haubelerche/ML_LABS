```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.datasets import ascent
```

```python
ascent_image = ascent()
```

```python

plt.gray()
plt.axis('off')
plt.imshow(ascent_image)
plt.show()
```

```python
temp_image = np.copy(ascent_image)
X = temp_image.shape[0]
Y = temp_image.shape[1]
print(f"image shape: {X,Y}")
```

```python
filter_=[[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]
weight = 1
```

```python
for x in range(1, X-1):
    for y in range(1,Y-1):
        conv = 0.0
        conv = conv + (ascent_image[x-1, y-1] * filter_[0][0])
        conv = conv + (ascent_image[x-1, y] * filter_[0][1])
        conv = conv + (ascent_image[x-1, y+1] * filter_[0][2])
        conv = conv + (ascent_image[x, y-1] * filter_[1][0])
        conv = conv + (ascent_image[x,y] * filter_[1][1])
        conv = conv + (ascent_image[x,y+1] * filter_[1][2])
        conv = conv + (ascent_image[x+1,y-1] * filter_[2][0])
        conv = conv + (ascent_image[x+1,y]* filter_[2][1])
        conv = conv + (ascent_image[x+1,y+1] * filter_[2][2])

        conv = conv * weight
        if conv < 0:
            conv = 0
        if conv > 255:
            conv = 255
        temp_image[x,y] = conv
```

```python
plt.gray()
plt.imshow(temp_image)
plt.show()
```

```python
#max_pooling
pool_x = int(X/2)
pool_y = int(Y/2)
pool_image = np.zeros((pool_x, pool_y))
for i in range(0,X, 2):  # bd filter
    for j in range(0,Y, 2):
        pixels= [temp_image[i, j], temp_image[i + 1, j], temp_image[i, j + 1], temp_image[i + 1,j + 1]]
        pool_image[int(i/2), int(j/2)] = max(pixels)

plt.gray()
plt.grid(False)
plt.imshow(pool_image)
plt.show()
```
